# Task
Create a bar chart comparing the world average and Southern Americas and Caribbean region average for the `GDP PPP`, `Youth Literacy Rate`, `Population (Urban %)`, and `Technology Exports` metrics, with appropriate labels, titles, and a legend.

## Calculate World Averages

### Subtask:
Calculate the mean of all numerical columns in the main DataFrame `df` (worldwide data). This will exclude non-numeric columns like 'Country/Territory', 'Code', and 'Region'.


**Reasoning**:
To calculate the world averages, I will first identify and select only the numerical columns from the main DataFrame `df`, excluding the specified non-numerical columns, and then compute the mean for each of these numerical columns, storing the result in `world_mean`.



```python
columns_to_exclude = ['Obs','Country/Territory', 'Code', 'Region']
numeric_df = df.drop(columns=columns_to_exclude, errors='ignore')
world_mean = numeric_df.mean(numeric_only=True)
print("World Averages (mean of numerical columns):")
display(world_mean)
```

## Calculate Region Averages

### Subtask:
Calculate the mean of the specified metrics (`GDP PPP`, `Youth Literacy Rate`, `Population (Urban %)`, and `Technology Exports`) for the 'Southern Americas and Caribbean' region using the `region_df` DataFrame.

#### Instructions
1. Select only the numerical columns from the `region_df` DataFrame.
2. Calculate the mean of each of the numerical columns for the region.
3. Store the resulting series of region averages in a variable named `region_mean`.

**Reasoning**:
To achieve the current subtask of calculating regional averages, I will first identify and select only the numerical columns from the `region_df` DataFrame and then compute the mean for each of these numerical columns, storing the result in `region_mean`.



```python
columns_to_exclude = ['Obs','Country/Territory', 'Code', 'Region']
reg_num = region_df.drop(columns=columns_to_exclude, errors='ignore')
region_mean = reg_num.mean(numeric_only=True)
print("Region Averages (mean of numerical columns):")
display(region_mean)
```

**Reasoning**:
The regional averages have been successfully calculated. Now, I will prepare a DataFrame specifically for the metrics requested in the main task, combining both world and regional averages, to facilitate plotting.



```python
metrics_to_plot = ['GDP PPP', 'Youth Literacy Rate', 'Population (Urban %)', 'Technology Exports']

# Create a DataFrame to hold the world and region means for the selected metrics
plot_df = pd.DataFrame({
    'world_mean': world_mean[metrics_to_plot],
    'region_mean': region_mean[metrics_to_plot]
})

# Calculate the difference for analysis (optional, but good for understanding)
plot_df['diff'] = plot_df['region_mean'] - plot_df['world_mean']

print("DataFrame for plotting (World vs. Region Averages):")
display(plot_df)
```

**Reasoning**:
To address the main task of comparing world and regional averages, I will create a bar chart for each of the specified metrics. Given the potentially disparate scales of these metrics, using subplots will ensure readability by allowing each metric to have its own y-axis scale and clear labeling.



```python
# Set style for plots
sns.set_style("whitegrid")

# Create subplots for each metric
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten() # Flatten the 2x2 array of axes for easy iteration

metrics_to_plot = ['GDP PPP', 'Youth Literacy Rate', 'Population (Urban %)', 'Technology Exports']
titles = {
    'GDP PPP': 'GDP PPP (USD)',
    'Youth Literacy Rate': 'Youth Literacy Rate (%)',
    'Population (Urban %)': 'Population (Urban %)',
    'Technology Exports': 'Technology Exports (Million USD)'
}

for i, metric in enumerate(metrics_to_plot):
    ax = axes[i]
    data_for_plot = plot_df.loc[metric, ['world_mean', 'region_mean']]
    data_for_plot.plot(kind='bar', ax=ax, color=['lightcoral', 'steelblue'])
    ax.set_title(f'Comparison of {titles[metric]}', fontsize=12)
    ax.set_ylabel(titles[metric], fontsize=10)
    ax.set_xlabel('') # Remove x-label as it's just 'world_mean', 'region_mean'
    ax.tick_params(axis='x', rotation=0) # Keep x-axis labels horizontal
    ax.legend(['World Average', 'Region Average'], loc='upper right')

plt.tight_layout()
plt.suptitle('World vs. Southern Americas and Caribbean Averages for Key Metrics', y=1.02, fontsize=16) # Overall title
plt.show()
plt.legend()
```
